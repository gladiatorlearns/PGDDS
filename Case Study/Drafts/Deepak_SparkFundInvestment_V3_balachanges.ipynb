```python
# Import the numpy and pandas packages

import numpy as np
import pandas as pd
```

```python
# Load the companies text file.

companies = pd.read_csv('companies.txt',encoding="ISO8859-1",sep="\t")

# Removing Special Characters from the 'permalink' column.
companies["permalink"]=companies["permalink"].apply(lambda x:x.encode('utf-8').decode('ascii','ignore'))

companies['name']=companies.name.str.encode('utf-8').str.decode('ascii','ignore')

# Varify the Data.
companies
```

```python
# Check the Records count.

companies.shape
companies.info()
```

```python
# Load the rounds2.csv file.
rounds2=pd.read_csv('rounds2.csv',encoding="ISO8859-1",sep=",")

# Remove the special characters from the 'Company_permalink' column.
rounds2["company_permalink"]=rounds2["company_permalink"].apply(lambda x:x.encode('utf-8').decode('ascii','ignore'))

rounds2
```

```python
# varify the records count...
rounds2.shape
```

```python
# How many unique companies are present in the companies file
companies.permalink.nunique()
```

```python
# How many unique companies are present in rounds2
rounds2.company_permalink.nunique()
```

```python
# check any company present in rounds2 not present in companies dataframe.

# make the column as lower case to compare..
rounds2['company_permalink'] = rounds2['company_permalink'].str.lower()

companies['permalink'] = companies['permalink'].str.lower()

# rounds2.loc[~rounds2['company_permalink'].isin(companies.permalink),:]

len(rounds2.loc[~rounds2['company_permalink'].isin(companies.permalink),:])
```

```python
# Check the columns having Null value ..
rounds2.isnull().sum()
```

```python
# Check the columns having Null value..
companies.isnull().sum()
```

```python
# merge dataframe to form master_frame.
master_frame=pd.merge(rounds2, companies,left_on="company_permalink" ,right_on="permalink",how="inner")
master_frame.head()
```

```python
# Number of observations/ Rows are present in master_frame
master_frame.shape
```

```python
# Check the columns having Null value ..
master_frame.isnull().sum()
```

```python
# Drop the rows having 'raised_amount_usd' having Null

master_frame=master_frame[~np.isnan(master_frame['raised_amount_usd'])]

master_frame.isnull().sum()
```

```python
# Records after Dropping Null records for 'raised_amount_usd'
master_frame.shape
```

```python
# convert the 'raised_amount_usd' to million dollar ..
master_frame[['raised_amount_usd']] = master_frame[['raised_amount_usd']].apply(lambda x: x / 1000000, axis=1)

master_frame
```

```python
# calculate the top Average value for the funding type.
master_frame.groupby('funding_round_type').raised_amount_usd.mean().sort_values(ascending=False)
```

```python
# calculate the top Average value for funding type with in a range of 5-15 M USD.
master_frame.groupby('funding_round_type').raised_amount_usd.mean().sort_values(ascending=False).between(5,15)
```

```python
# get the Top 9 countries based on theÂ total investment amount each country has received .
# consider the funding type as 'venture' (invest between 5 to 15 million USD)

#top9 =master_frame.loc[(master_frame.funding_round_type =='venture')].sort_values(by = (['raised_amount_usd']), ascending = False)
#top9.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False).head(9)

top9 =master_frame.loc[(master_frame.funding_round_type =='venture')].groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False).head(9)
top9

## Top english speaking country
# USA,GBR,IND (as CHN is not the list)
```

```python
## extract and create the primary sector from the Category list..
master_frame['primary_sector']=master_frame.category_list.str.split('|', expand = True)[0]
master_frame
```

```python
## load mapping csv file.

mapping=pd.read_csv('mapping.csv',na_filter=False)

##It is observed that some places data having wrong value .replace those data..
mapping['category_list']=mapping['category_list'].apply(lambda x: x.replace('0', 'na'))

mapping
```

```python
## Reshape the mapping dataframe..
mapping_temp=pd.melt(mapping,id_vars=["category_list"],var_name='main_sector')
mapping_temp

## fitering the value=1 columns only ()
mapping=mapping_temp[mapping_temp['value']==1]
mapping
```

```python
## Prepare the master frame with main sector mapping..

master_frame_sector=pd.merge(master_frame, mapping[['main_sector','category_list']],left_on="primary_sector" ,right_on="category_list",how="inner")
master_frame_sector
```

```python
#master_frame_sector.info()

master_frame_sector=master_frame_sector.drop(['category_list_y'], axis=1)
master_frame_sector
```

```python
master_frame_sector.isnull().sum()
```

```python
## find out the most heavilyÂ invested main sectors in each of the three countries (USA,GBR,IND)
## (for funding type FT and investments range of 5-15 M USD)

# D1=master_frame_sector[(master_frame_sector['country_code']=='USA') & (master_frame_sector['funding_round_type']=='venture')]
# D1.shape  #(35784, 18)

D1=master_frame_sector[(master_frame_sector['country_code']=='USA') & (master_frame_sector['funding_round_type']=='venture') & (master_frame_sector['raised_amount_usd'] >=5) & (master_frame_sector['raised_amount_usd'] <=15 )]
D1.shape #(12012, 18)
D1
```

```python
## New column for the total number (or count) of investments for each main sector 
D1['tot_investment_per_sector'] = D1['company_permalink'].groupby(D1['main_sector']).transform('count')

##  New Column for the total amount invested in each main sector 
D1['tot_amountinvested_per_sector'] = D1['raised_amount_usd'].groupby(D1['main_sector']).transform('sum')

D1
```

```python
# 1 A plot showing the fraction of total investments (globally) in venture, seed, and private equity, 
# and the average amount of investment in each funding type

df_sector=master_frame_sector.loc[master_frame_sector['funding_round_type'].isin(['venture','seed','private_equity'])]

plot1_totinvestment = df_sector.groupby('funding_round_type').agg({"raised_amount_usd": ['sum','mean']}) 
plot1_totinvestment.columns = ["_".join(x) for x in plot1_totinvestment.columns.ravel()]
plot1_totinvestment
```

```python
# 2.A plot showing the top 9 countries against the total amount of investments of funding type FT

plot2_top9countries =df_sector.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False).head(9)
plot2_top9countries
```
